```python
import math
```

```python
logits_1 = [1.9, -2.1, -0.3, 0.4, -0.8]
logits_2 = [0.3, 2.5, -0.5, 0.0, -1.1]
logits_3 = [-0.6, 0.2, 1.4, -0.9, -0.2]
```

```python
def softmax(x):
    p = []
    for i in range(len(x)):
        p.append(math.exp(x[i]))
    total = sum(p)
    for i in range(len(x)):
        p[i] = p[i] / total
    return p
```

```python
probs1 = softmax(logits_1)
print(probs1)
print(sum(probs1))
```

```python
L1 = -math.log(probs1[1])
print(L1)
```

```python
probs2 = softmax(logits_2)
print(probs2)
print(sum(probs2))
```

```python
L2 = -math.log(probs2[0])
print(L2)
```

```python
probs3 = softmax(logits_3)
print(probs3)
print(sum(probs3))
```

```python
L3 = -math.log(probs3[2])
print(L3)
```

```python
L_total = L1 + L2 + L3
Average_Loss = L_total / 3
print(Average_Loss)
```
